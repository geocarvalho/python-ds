* Correlation between attributes

 * Correlation refers to the relationship between two variables and how they may or may not change together;

 * The Pearson's correlation coefficient method assumes a normal distribution of the attributes involved;

 * A correlation of -1 or 1 shows a full negative or positive correlation respectively. Whereas a value of 0 shows no correlation at all;

 * Some ML algorithms like linear and logistic regression can suffer poor performance if there are highly correlated attributes in your dataset.

```
# Correlation between attributes
data.corr(method='pearson')
```

* Skew of univariate Distributions
 * Refers to a distribution that is assumed Gaussian (normal or bell curve) that is shifted or squashed in one direction or another;

 * Many ML algorithms assume a Gaussian distribution, sometimes is necessary perform data preparation to correct **skew** and later improve the accuracy of the model;

 * The skew results are positive or negative, values closer to zero show less skew.

```
# Skew of univariate Distributions
data.skew()
```

## Chapter 6 - Understand your data with visualization

* The fastest way to learn more about your data is to use data visualization

### Histograms

* Histograms group data into bins and provide you a count of the number of observations in each bin;

* From the shape of the bins you can quickly get a feeling for whether an attribute is Gaussian, skewed or even has an exponential distribution;

* It can also help you see possible outliers.


```
from matplotlib import pyplot as plt
%matplotlib inline

data.hist(figsize=(8,8))
plt.show()
```

* We can see that perhaps the attributes `age`, `pedi` and `test` may have an **exponential distribution**;

* We can also see that perhaps the `mass`, `pres` and `plas` attributes may have nearly a **Gaussian distribution**;

> Many ML techiniques assume a Gaussian univariate distribution on the input variable.

### Density plots

* The plots look like an abstracted histogram with a smooth curve drawn through the top of each bin.

```
data.plot(kind='density' ,subplots=True, layout=(3,3), sharex=False, figsize=(15,15))
plt.show()
```

### Box and whisker plots

* Boxplots summarize the distribution of each attribute, drawing a line for the median (middle value) and a box around the 25th and 75th percentiles (the middle 50% of the data);

* The whiskers give an idea of the spread of the data and dots outside of the whiskers show candidate outlier values (values that are 1.5 times greater than the size of spread of the middle 50% of the data).

```
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(9,9))
plt.show()
```

### Multivariate plots

#### a) correlation matrix plot

* Correlation matrix gives an indication of how related the changes are between two variables;

* If two variables change in the same direction they are **positively correlated** (near +1). If they change in opposite directions together (one goes up, one goes down), then they are **negatively correlated** (near -1);

* Some ML algorithms (linear and logistic regression) can have poor performance if there are highly correlated input variables in the dataset.

```
correlations = data.corr()
# plot correlation matrix
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

#### b) scatter plot matrix

* It shows the relationship between two variables as dots in two dimensions, one axis for each attribute;

* Drawing all these scatter plots together is called a scatter plot matrix;

* Attributes with structured relationships may also be correlated and good candidates for removal from your dataset.

```
pd.plotting.scatter_matrix(data, figsize=(20,20))
plt.show()
```

## Chapter 7 - Prepare your data for ML

### Rescale data

* When your data is comprised of attributes with varying scales, many ML algorithms can benefit from rescaling the attributes to all have the same scale;

> Often this is referred to as **normalization** and attributes are often rescaled into the range between 0 and 1.

* This is useful for optimization algorithms used in the core of ML algorithms like *gradient descent*;

* It's also useful for algorithms that weight inputs like regression and NNs, and algorithms that use distance measures (k-NN);

* You can rescale your data using scikit-learn `MinMaxScaler` class:

```
  from sklearn.preprocessing import MinMaxScaler

array = data.values
# separate array into input and output components
X = array[:,0:8]
Y = array[:,8]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
# summarize transformed data
np.set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

### Standardize data

* Standardization is a useful technique to transform attributes with a Gaussian distribution and differing means and standard deviations to a standard *Gaussian distribution* with a *mean* of 0 and a *standard deviation* of 1;

* It's most suitable for techniques that assume *Gaussian distribution* in the input variables and work better with rescaled data (linear regression, logistic regression, linear discriminate analysis);

* We can standardize data using scikit-learn `StandarScaler` class:

> The values for each attribute now have a mean value of 0 and a standard deviation of 1.

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
# summarize transformed data
np.set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

### Normalize data

* Normalizing in scikit-learn refers to rescaling each observation (row) to have a length of 1 (called a unit norm or a vector with the length of 1 in linear algebra);

* This pre-processing method can be useful for sparse datasets (lots of zeros) with attributes of varying scales when using algorithms that weight input values (NNs) and algorithms that use distance measures (k-NN);

* We can normalize data with scikit-learn using the `Normalizer` class:

> The rows are normalized to length 1.

```
# Normalize data (length of 1)
from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
# summarize transformed data
np.set_printoptions(precision=3)
print(normalizedX[0:5,:])
```

### Binarize data

* All values above the threshold are marked 1 and all equal to or below are marked as 0;

* It can be useful when you have probabilities that you want to make crisp values;

* It's also useful when feature engineering and you want to add new features that indicate something meaningful;

* We can create new binary attributes using scikit-learn `Binarizer` class:


```
# binarization
from sklearn.preprocessing import Binarizer

binarizer = Binarizer(threshold=0.0).fit(X)
binaryX = binarizer.transform(X)
# summarize transformed data
np.set_printoptions(precision=3)
print(binaryX[0:5,:])
```
