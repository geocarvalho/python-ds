```python
import pandas as pd
from pandas_datareader import data
```

```python
companies = ['MSFT', 'GOOG', 'AAPL', 'YHOO', 'AMZN']
p = data.DataReader(name=companies, data_source='yahoo',
                   start='2010-01-01', end='2016-12-31')
p
```

# Transpose a `Panel` with the `.transpose()` method

```python
p.axes
```

```python
p2 = p.transpose(2,1,0) # reorganize axes
```

```python
p2.axes
```

```python
p2['AAPL']
```

```python
p2.major_xs(p2.major_axis[1])
```

```python
p2.minor_xs('Volume')
```

# The `.swapaxes()` method

```python
p2 = p.swapaxes('items', 'minor')
p2.axes
```

```python
p2['MSFT']
```

```python
p2.major_xs(p2.major_axis[2])
```

```python
p2.minor_xs('Close')
```
