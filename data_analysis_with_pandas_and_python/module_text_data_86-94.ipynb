# The `expand` and `n` parameters ot the `.str.split()` method

```python
chicago = pd.read_csv('/home/george/Documents/Cursos/data_analysis_with_pandas_and_python/pandas/chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.head()
```

```python
chicago[['First Name', 'Last Name']] = chicago['Name'].str.split(',', expand=True)
chicago.head()
```

```python
chicago[['First Title Word', 'Remaining Words']] = chicago['Position Title'].str.split(' ', expand=True, n=1)
```
